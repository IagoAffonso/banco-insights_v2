```python
# Create a json file with the unique values of the columns NomeRelatorio, Grupo, NomeColuna, DescricaoColuna,NomeRelatorio_Grupo_Coluna, and save it in the current directory
import pandas as pd
import json

df = pd.read_csv('../bacen_project_v1/data/consolidated_cleaned.csv')

unique_values = {
    'NomeRelatorio': df['NomeRelatorio'].unique().tolist(),
    'Grupo': df['Grupo'].unique().tolist(),
    'NomeColuna': df['NomeColuna'].unique().tolist(),
    'DescricaoColuna': df['DescricaoColuna'].unique().tolist(),
    'NomeRelatorio_Grupo_Coluna': df['NomeRelatorio_Grupo_Coluna'].unique().tolist()
}

with open('unique_values.json', 'w', encoding='utf-8') as f:
    json.dump(unique_values, f, ensure_ascii=False, indent=2)
```
